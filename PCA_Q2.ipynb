##Libraries

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

##Load Image

```python
# Load the grayscale image
image_path = '/content/student_gray.jpg'
image = Image.open(image_path).convert('L')

image_np = np.array(image)
plt.imshow(image_np, cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.show()

# Shape of the image
image_np.shape
```

```python
# Reshape the image into a 2D matrix
reshaped = image_np.reshape(-1, image_np.shape[1])
```

##PCA function

```python

def pca_and_reconstruct(n_components, original):
    pca = PCA(n_components=n_components)
    transformed = pca.fit_transform(original)
    reconstructed = pca.inverse_transform(transformed)
    return reconstructed

components = [5, 20, 50, 100]
```

##Reconstruction & Visualisation

```python
fig, axs = plt.subplots(1, len(components)+1, figsize=(15, 5))

# Original image
axs[0].imshow(image_np, cmap='gray')
axs[0].set_title("Original")
axs[0].axis('off')

# Reconstructed images
for i, n in enumerate(components):
    reconstructed = pca_and_reconstruct(n, reshaped)
    axs[i+1].imshow(reconstructed.reshape(image_np.shape), cmap='gray')
    axs[i+1].set_title(f"{n} components")
    axs[i+1].axis('off')

plt.tight_layout()
plt.show()
```
